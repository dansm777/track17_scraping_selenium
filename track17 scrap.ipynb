```python

for i in range(10):
    print(i)
    data=[]
    data.append(file.iloc[i][0].lstrip())
    data.append(file.iloc[i][1].lstrip())
    data.append(file.iloc[i][2])
    data.append(file.iloc[i][3])
    data.append(file.iloc[i][4])
    data.append(file.iloc[i][5])
    data.append(file.iloc[i][6])
    data.append(file.iloc[i][7])
    data.append(file.iloc[i][8])
    data.append(file.iloc[i][9])
    data.append(file.iloc[i][10])
    data.append(file.iloc[i][11])
    #data.append(file.iloc[i][12])
    track=""
    trck=file.iloc[i][3] 
    trck=re.sub('\s+', '', trck)
    print(trck.lstrip())
    curl=url+str(trck.lstrip())
    print(curl)
    try:
        browser.get(curl)
        time.sleep(3)
        table_id = browser.find_element_by_xpath("/html[1]/body[1]/main[1]/div[1]/article[1]/div[2]/div[1]/div[1]/div[1]/div[2]/div[2]/div[1]/dl[2]")
        rows = table_id.find_elements_by_tag_name("dd")
        for row in rows:
            col = row.find_elements_by_tag_name("p")[0]
            #print (col.text)
            track+=col.text+','
    except:
        try:
            t=browser.find_element_by_xpath("/html/body/main/div/article/div[2]/div[1]/div/div/div[2]/div[1]/div[2]")
            if (t.text=='Select carrier'):
                if (file.iloc[i][3]=='USPS'):

                    try:
                        t2=browser.find_element_by_xpath("/html/body/main/div/article/div[2]/div[1]/div/div/div[2]/div[2]/div/div/a[1]")
                        t2.click()
                        time.sleep(5)
                        table_id = browser.find_element_by_xpath("/html[1]/body[1]/main[1]/div[1]/article[1]/div[2]/div[1]/div[1]/div[1]/div[2]/div[2]/div[1]/dl[2]")
                        rows = table_id.find_elements_by_tag_name("dd")
                        for row in rows:
                            col = row.find_elements_by_tag_name("p")[0]
                            #print (col.text)
                            track+=col.text+','
                    except:
                        print("not available")
                        track+="not available"
                elif (file.iloc[i][3]=='FEDEX'):
                    #print("fed")


                    try:
                        t2=browser.find_element_by_xpath("/html/body/main/div/article/div[2]/div[1]/div/div/div[2]/div[2]")
                        cnt=int((t2.text.split(":")[1].split("\n")[1:].index('Fedex'))/2)+1
                        #print(cnt)
                        t3=browser.find_element_by_xpath("/html/body/main/div/article/div[2]/div[1]/div/div/div[2]/div[2]/div/div/a["+str(cnt)+"]")         
                        #print('i')                       
                        t3.click()
                        time.sleep(5)
                        table_id = browser.find_element_by_xpath("/html[1]/body[1]/main[1]/div[1]/article[1]/div[2]/div[1]/div[1]/div[1]/div[2]/div[2]/div[1]/dl[2]")
                        rows = table_id.find_elements_by_tag_name("dd")
                        for row in rows:
                            col = row.find_elements_by_tag_name("p")[0]
                            #print (col.text)
                            track+=col.text+','
                    except:
                        print("not available")
                        track+="not available"

                else:
                    print("not available")
                    track+="not available"

            else:
                print("not available")
                track+="not available"

        except:
            print("not available")
            track+="not available"
    
    data.append(track)
     
    if (('return' in track )or ( 'Return' in track )or ( 'returned' in track )or
        ( 'Returned' in track )or ( 'returning' in track )or ( 'Returning' in track)):
        data.append(1)
    else:    
        data.append(0)
    
    if (('shipper' in track) or ('Shipper' in track) or ('Original Sender') or ('sender' in track)
       or ('Sender' in track)):
        data.append(1)
    else:
        data.append(0)
    print(data)
    final_data.append(data)
    print("********************************************")
```
